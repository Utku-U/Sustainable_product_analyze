## For Example

```python
import pandas as pd
import numpy as np

np.random.random(42)

data = {
    'Yas': np.random.randint(18, 60, 20),           # 1. Değişken
    'Yillik_Gelir': np.random.randint(30000, 150000, 20), # 2. Değişken
    'Sitede_Gecen_Sure_Dk': np.random.randint(1, 45, 20), # 3. Değişken
    'Toplam_Harcama': np.random.uniform(100.0, 5000.0, 20).round(2), # 4. Değişken

    'Meslek': np.random.choice(['Mühendis', 'Doktor', 'Öğrenci', 'Sanatçı'], 20), # 5. Değişken
    'Sehir': np.random.choice(['İstanbul', 'Ankara', 'İzmir', 'Bursa'], 20),      # 6. Değişken
    'Cihaz_Tipi': np.random.choice(['Mobil', 'Masaüstü', 'Tablet'], 20),          # 7. Değişken

    # 1: Tıkladı, 0: Tıklamadı
    'Tiklama_Durumu': np.random.randint(0, 2, 20)   # 8. Değişken (Target)
}

df = pd.DataFrame(data)

df.head() 
```

### Data Preposition

```python
deep_features = df[["Yas", "Yillik_Gelir", "Sitede_Gecen_Sure_Dk", "Toplam_Harcama"]].values
wide_features = pd.get_dummies(df[["Meslek","Sehir","Cihaz_Tipi"]]).values
y_target = df[["Tiklama_Durumu"]].values

# Modelimizi oluştururken modele girecek input shape sayısını belirlemek için kendi modelimi referans alıyoruz. Yani bu modeli kullanacak olanlar
# 4 değişkenlerini deep inputundan, 11 değişkenlerini de wide inputundan sokacaklar gibi...
col_wide_features = wide_features.shape[1] # Sütun sayısı 11
col_deep_features = deep_features.shape[1] # Sütun sayısı 4
```

### Making model

```python
# MODEL OLUŞTURMA AŞAMASI

# !pip install scikit-learn
# !pip install tensorflow==2.20.0 

from tensorflow.keras import Input, layers, Model
import keras

Input_wide = keras.layers.Input(shape=(col_wide_features,), name="wide_input") # Modelin memorize edeceği değşler.
Input_deep = keras.layers.Input(shape=(col_deep_features,), name="deep_input") # Modelin generalize edeceği değşler.
hidden1 = keras.layers.Dense(30, activation="relu")(Input_deep)
hidden2 = keras.layers.Dense(20, activation="relu")(hidden1)
concat = keras.layers.concatenate([Input_wide, hidden2])
output = keras.layers.Dense(1, activation="sigmoid", name="output")(concat)

model = keras.Model(inputs=[Input_wide, Input_deep], outputs=[output]) 
```

### Building Optimizer,loss,metrics that we maked model 

```python
from tensorflow import optimizers
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report

model.compile(loss="binary_crossentropy", optimizer=keras.optimizers.SGD(learning_rate=1e-3), metrics=["accuracy", 
                                                                                                       keras.metrics.Precision(name='precision'),
                                                                                                       keras.metrics.Recall(name='recall'),
                                                                                                       keras.metrics.F1Score(name='F1')]) 
```

## Train Test Validation

```python
train_size = int(len(df)*0.80)

X_train_wide = wide_features[:train_size] 
X_test_wide = wide_features[train_size:]

X_train_deep = deep_features[:train_size]
X_test_deep = deep_features[train_size:]

y_train = y_target[:train_size]
y_test  = y_target[train_size:] 
```

```python
history = model.fit((X_train_wide, X_train_deep), y_train, epochs=10) 
```

```python
X_new_wide, X_new_deep = X_test_wide, X_test_deep

y_pred_prob = model.predict([X_new_wide, X_new_deep])
```

```python
y_pred_class = (y_pred_prob > 0.5).astype("int32")
y_pred_class, y_test 
```

## Class of WideandDeepModel

```python
from tensorflow.keras import Input, layers, Model
import keras
from tensorflow import optimizers
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report


class WideAndDeepModel(keras.Model):
    
    def __init__(self, units=30, activation="relu", **kwargs):
        super().__init__(**kwargs) 
        self.hidden1 = keras.layers.Dense(units, activation=activation)
        self.hidden2 = keras.layers.Dense(units, activation=activation)
        self.main_output = keras.layers.Dense(1)
        self.aux_output = keras.layers.Dense(1)
        
    def call(self, inputs):
        input_A, input_B = inputs
        hidden1 = self.hidden1(input_B)
        hidden2 = self.hidden2(hidden1)
        concat = keras.layers.concatenate([input_A, hidden2])
        main_output = self.main_output(concat)
        aux_output = self.aux_output(hidden2)
        return main_output, aux_output    

model = WideAndDeepModel() 
model.call
```
